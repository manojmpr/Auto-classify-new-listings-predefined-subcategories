```python
city_list = city_list = ['dallas', "houston", "chicago", "phoenix", "newyork", "philadelphia", "sandiego","losangeles", "sanantonio", "sfbay"]
links = []
for city in city_list:
  page = requests.get(f"https://{city}.craigslist.org/d/appliances/search/ppa")
  bsobj = bs4(page.content,'lxml')
  for link in bsobj.findAll('li',{'class':'result-row'}):
    sublink=link.find('a')
    html_class=sublink.get('href')
    links.append(html_class)
```

```python
def getdata(url): 
    r = requests.get(url) 
    return r.text 
```

```python
img=[]
for i in links:
    htmldata=getdata(i)
    soup = BeautifulSoup(htmldata, 'html.parser') 
    for item in soup.find_all('img'):
        img.append(item['src'])
```
